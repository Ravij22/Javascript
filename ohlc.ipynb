```python
import json
with open('C:\hackathon\Stock_List.json','r') as f:
    data = json.load(f)

c=[]
h=[]
l=[]
o=[]
s=[]
date=[]
i=0

sym = input("Enter symbol:")

for i in range(len(data)):
    if data[i]['symbol']==sym:
        c.append(data[i]['close'])
        h.append(data[i]['high'])
        l.append(data[i]['low'])
        o.append(data[i]['open'])
        s.append(data[i]['symbol'])
        date.append(data[i]['date'])
    
print(data[0])

```

```python
import plotly.graph_objects as go
```

```python
fig = go.Figure(data=[go.Candlestick(x=date,open=o,high=h,low=l,close=c)])
                      
fig.show()
```
